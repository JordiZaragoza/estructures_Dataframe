### *Descarga el data set Airlines Delay: Airline on-time statistics and delay causes i cargalo en un pandas Dataframe. Explora los datos que contiene, y quedate unicamente con las columnas que consideres relevantes*

```python
import pandas as pd

df = pd.read_csv("DelayedFlights.csv") # cargo el archivo csv
df.head()
```

#### Columnas a usar

```python
df = df[["Year","Month","DayofMonth","DepTime","CRSDepTime","ArrTime","CRSArrTime","UniqueCarrier","AirTime","ArrDelay","DepDelay","Distance","Origin","Dest"]]
df.head()
```

### *Crea columnas nuevas (velocidad media de vuelo, si ha llegado tarde o no...)*

```python
df['AvgSpeed'] = round(df['Distance']/(df['AirTime']/60)) # velocidad media
df['ArrLate'] = df['ArrDelay']>0 # si ha llegado tarde o no
df.head()
```

### *Tabla las aerolineas con mas retrasos acumulados*

```python
df2 = df.groupby("UniqueCarrier").ArrLate.sum("True") # agrupo las aerolineas y sumo los retrasos totales
df2 = df2.sort_values(ascending=False) # ordeno de manera descendente
df2.head(10) # muestro las 10 aerolineas con mas retrasos
```
